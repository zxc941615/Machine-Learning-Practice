```python
import pandas as pd
import numpy as np
```

```python
# read file
stock_path = "C:/Users/acer/Desktop/LAB/lab2_2_2.csv"
stock_df = pd.read_csv(stock_path)
```

```python
stock_df
```

```python
stock_df = stock_df[5:]
```

```python
stock_df
```

```python
K = [0]
D = [0]
SMA = [0]
EMA = [0]
RSI = [0]

for i in range(1,len(stock_df)):
    if(stock_df[i:i+1]["K"].values <=80 and stock_df[i:i+1]["K"].values>=20):
        if(stock_df[i:i+1]["K"].values > stock_df[i-1:i]["K"].values):
            K.append(1)
        else:
            K.append(-1)
    elif(stock_df[i:i+1]["K"].values>80):
        K.append(-1)
    else:
        K.append(1)
        
        
    if(stock_df[i:i+1]["D"].values <=80 and stock_df[i:i+1]["D"].values>=20):
        if(stock_df[i:i+1]["D"].values > stock_df[i-1:i]["D"].values):
            D.append(1)
        else:
            D.append(-1)
    elif(stock_df[i:i+1]["D"].values>80):
        D.append(-1)
    else:
        D.append(1)    
        
    if(stock_df[i:i+1]["SMA"].values > stock_df[i-1:i]["SMA"].values):
        SMA.append(1)
    else:
        SMA.append(-1)
    
    if(stock_df[i:i+1]["EMA"].values > stock_df[i-1:i]["EMA"].values):
        EMA.append(1)
    else:
        EMA.append(-1)
    
    if(stock_df[i:i+1]["RSI"].values <=80 and stock_df[i:i+1]["RSI"].values>=20):
        if(stock_df[i:i+1]["RSI"].values > stock_df[i-1:i]["RSI"].values):
            RSI.append(1)
        else:
            RSI.append(-1)
    elif(stock_df[i:i+1]["RSI"].values>80):
        RSI.append(-1)
    else:
        RSI.append(1)
```

```python
len(K)
```

```python
stock_df["KR"] = K
stock_df["DR"] = D
stock_df["SMAR"] = SMA
stock_df["EMAR"] = EMA
stock_df["RSIR"] = RSI
```

```python
stock_df.to_csv(paindex = False)
```
