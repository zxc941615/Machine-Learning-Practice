```python
import pandas as pd
import numpy as np
import torch
import torch.utils.data as Data
from torch.autograd import Variable
```

```python
stock_path = "C:/Users/acer/Desktop/LAB/3450Inf.csv"
stock_df = pd.read_csv(stock_path)
stock_df['Date'] = pd.to_datetime(stock_df['Date'])
stock_df.index = stock_df["Date"]
stock_df
```

```python
df = stock_df[7:2407][["Close","Change(%)","market_Change(%)","Beta_7D","K value","D value","MACD","William"]]
train_set = df[:2000]
test_set = df[2000:]
print(train_set.shape)
print(test_set.shape)
```

```python
features = []
for i in range(7,2000):
    x = train_set[i-7:i][["Close","Change(%)","market_Change(%)","Beta_7D","K value","D value","MACD","William"]].values
    features.append(x.tolist())
features = torch.FloatTensor(features)

labels = []
for i in range(8,2001):
    x = train_set[i-1:i]["Close"]
    labels.append(x.tolist())
labels = torch.FloatTensor(labels)
```

```python
print(features[0])
print(labels[0])
```

```python
print(features.size())
print(labels.size())
labels = labels.view(-1)
print(labels.size())
```

```python
EPOCH = 50
BATCH_SIZE = 100
TIME_STEP = 7
INPUT_SIZE = 8
LR = 0.01
```

```python
class RNN(torch.nn.Module):
    def __init__(self):
        super(RNN,self).__init__()
        self.lstm = torch.nn.LSTM(
            input_size=INPUT_SIZE,
            hidden_size=64,         
            num_layers=5,
            dropout=0.5,
            batch_first=True, 
        )
        self.out = torch.nn.Linear(64,1)
        
    def forward(self,x):
        r_out,(h_n,h_c) = self.lstm(x,None)
        out = self.out(r_out[:,-1,:])
        return out
```

```python
model = RNN()
print(model)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=LR)   
loss_func = torch.nn.MSELoss()
```

```python
LOSS = []
loss_total = 0
for epoch in range(EPOCH):
    inputs = Variable(features)
    target = Variable(labels)
    output = model(inputs)
    output = output.view(-1)
    loss = loss_func(output,target)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    loss_total = loss_total + loss
    if((epoch+1)%10 == 0):
        avg = loss_total / 10
        LOSS.append(avg.tolist)
        print("Epoch: %d | Train Avg Loss: %.4f "%(epoch+1,avg))
        loss_total = 0
            
```
