# 資料若為字串的話,如果要訓練模型,則需要轉為數值 ,方式有兩種
    label enconding : 若資料為有序的 如 小孩 年輕人 老人  表示為 0 1 2 是合理的
    one hot enconding : 使用時機為資料是無序的 如國家

```python
import numpy as np 
import pandas as pd
```

```python
path = './data/application_train.csv'
app_train = pd.read_csv(path)
app_train.head()
```

# Using One hot enconfing

```python
sub_train = pd.DataFrame(app_train['WEEKDAY_APPR_PROCESS_START'])
print(sub_train.shape)
sub_train
```

```python
sub_train = pd.get_dummies(sub_train)
print(sub_train.shape)
sub_train.head()
```
