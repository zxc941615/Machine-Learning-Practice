# GET DATA FROM WEB

```python
import requests
url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
response = requests.get(url)
data = response.text
```

```python
print(len(data))
```

```python
data[0:100]
```

```python
data = data.split('\n')
print(len(data))
```

```python
data[0:5]
```

```python
buffer = []
for i in data:
    line  = i.split("\t")
    buffer.append(line)
df = pd.DataFrame(buffer)
df
```

# use pd.read_csv

```python
df2 = pd.read_csv(url,sep = '\t',header = None)
df2
```

```python
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
##  read image
response = requests.get(df2.loc[0,1])
img = Image.open(BytesIO(response.content))
img = np.array(img)
print(img.shape)
plt.imshow(img)
plt.show
```

```python
def getImageFromURL(url_list):
    list = []
    for url in url_list:
        response = requests.get(url)
        
        try:
            img = Image.open(BytesIO(response.content))
            img = img.resize((255,255))
            img = np.array(img)
            list.append(img)
        except:
            pass
    return list
```

```python
result = getImageFromURL(df2[0:5][1].values)
print(len(result))
```

```python
for i in result:
    plt.imshow(i)
    plt.show()
```
